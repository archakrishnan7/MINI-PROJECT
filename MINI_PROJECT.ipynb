```python
import os
import time
from datetime import datetime
import numpy as np
import pandas as pd
```

```python
v_data = pd.read_csv("/content/Electric_Vehicle_Population_Data.csv")
out_dir = "outputs"
intervel_seconds = 20
max_runs = 3
```

```python
def extract():
  v_data = pd.read_csv('/content/Electric_Vehicle_Population_Data.csv')
  v_data_len = len(v_data)
  print(f"Extracted {v_data_len}rows")
  return v_data

def transform(v_data):

  v_data = v_data.drop_duplicates()

  print(v_data.isnull().sum())


  numeric_col = v_data.select_dtypes(include=['int64','float64']).columns
  v_data[numeric_col] = v_data[numeric_col].fillna(v_data[numeric_col].median())
  text_col = v_data.select_dtypes(include='object').columns
  v_data[text_col] = v_data[text_col].fillna("Unknown")


  v_data["Postal Code"] = v_data["Postal Code"].astype("int64")

  v_data["Vehicle Age"] = 2026 - v_data["Model Year"]

  coordinates = v_data["Vehicle Location"].str.extract(r'POINT \(([-\d.]+) ([-\d.]+)\)')
  v_data["Longitude"] = coordinates[0].astype(float)
  v_data["Latitude"] = coordinates[1].astype(float)

  v_data.reset_index(drop=True,inplace=True)

  return v_data

def load(v_data):
    os.makedirs(out_dir, exist_ok=True)

    output_path = f"{out_dir}Electric_Vehicle_Population_Data.csv"

    v_data.to_csv(output_path, index=False)

    print(f"Loaded {len(v_data)} rows")
    print(f"Saved transformed data to: {output_path}")







```

```python
v_data = extract()
v_data = transform(v_data)
load(v_data)
```
